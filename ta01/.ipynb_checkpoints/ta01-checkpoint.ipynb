Importa bibliotecas que serão usadas

```python
import numpy as np
import cv2
import sys
import os
from matplotlib import pyplot as plt
```

Lê e binariza a imagem aplicando técnica de threshold

```python
#Lê imagem
img = cv2.imread("./teste.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Gera imagem binária usando métrica de Otsu
_, bin_img = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)

plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(bin_img, 'gray')
plt.show()
```

Aplica operação morfologica de dilatação, assim juntando as letras para transformar a palavra em um bloco

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
b_rect = cv2.morphologyEx(bin_img, cv2.MORPH_GRADIENT, kernel)
b_rect = cv2.dilate(b_rect,kernel,iterations = 3)
```

Procura os contornos e desenha na imagem um retângulo em volta da palavra

```python
cont = 0
img2 = cv2.imread("./teste.jpg")
#Acha os contornos na imagem dilatada
cnt, _ = cv2.findContours(b_rect, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 
for c in cnt:
    perim = cv2.arcLength(c, True)
    if perim > 150: #Se o contorno for grande o suficiente, classifica como palavra
        cont += 1
        x,y,w,h = cv2.boundingRect(c)
        cv2.rectangle(img2,(x,y),(x+w,y+h),(0,0,255),2)

plt.figure(figsize=(15,10))
plt.imshow(img2)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.show()
print("Número de palavras: ",cont)
```

O resultado anterior foi feito utilizando a imagem sem aplicar nenhum filtro. Agora iremos aplicar o filtro pela média na imagem binarizada a fim de tirar ruidos e não confundi-los com palavras

```python
#Aplica o filtro pela média
bin_img_filtered = cv2.medianBlur(bin_img,5)

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
b_rect = cv2.morphologyEx(bin_img_filtered, cv2.MORPH_GRADIENT, kernel)
b_rect = cv2.dilate(b_rect,kernel,iterations = 3)

cont = 0
#Acha os contornos na imagem dilatada
cnt, _ = cv2.findContours(b_rect, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 
for c in cnt:
    perim = cv2.arcLength(c, True)
    if perim > 150: #Se o contorno for grande o suficiente, classifica como palavra
        cont += 1
        x,y,w,h = cv2.boundingRect(c)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)

plt.figure(figsize=(15,10))
plt.imshow(img)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.show()
print("Número de palavras: ",cont)
```
